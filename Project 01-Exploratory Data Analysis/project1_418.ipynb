```python
# load libraries 
import pandas as pd
pd.set_option('display.max_columns',100)
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
# load the dataset using pandas.read_csv
DEMOGRAPHICS_DATA = pd.read_csv('demographics_train.csv')
ELECTIONS_DATA = pd.read_csv('election_train.csv')
```

```python
# TASK 1 - delete the word " County" and " Parish" from the county column
DEMOGRAPHICS_DATA['County'] = DEMOGRAPHICS_DATA['County'].str.replace(' County','')
DEMOGRAPHICS_DATA['County'] = DEMOGRAPHICS_DATA['County'].str.replace(' Parish','')
DEMOGRAPHICS_DATA.head()
```

```python
# TASK 1 CONTINUED - change state names to abbreviated names  
abbreviated = {'Alabama':'AL','Alaska':'AK','Arizona':'AZ','Arkansas':'AR','California':'CA','Colorado':'CO',
               'Connecticut': 'CT','Delaware': 'DE','Florida': 'FL','Georgia':'GA','Hawaii':'HI','Idaho':'ID',
               'Illinois':'IL','Indiana':'IN','Iowa':'IA','Kansas':'KS','Kentucky':'KY','Louisiana':'LA',
               'Maine':'ME','Maryland':'MD','Massachusetts':'MA','Michigan':'MI','Minnesota':'MN','Mississippi':'MS',
               'Missouri':'MO','Montana':'MT','Nebraska':'NE','Nevada':'NV','New Hampshire':'NH','New Jersey': 'NJ',
               'New Mexico':'NM','New York':'NY','North Carolina':'NC','North Dakota':'ND','Ohio':'OH','Oklahoma':'OK',
               'Oregon': 'OR','Pennsylvania':'PA','Rhode Island': 'RI','South Carolina':'SC','South Dakota': 'SD',
               'Tennessee':'TN','Texas':'TX','Utah':'UT','Vermont':'VT','Virginia':'VA','Washington':'WA',
               'West Virginia':'WV','Wisconsin':'WI','Wyoming':'WY'}

ELECTIONS_DATA['State'] = ELECTIONS_DATA['State'].map(abbreviated)
```

```python
# TASK 1 CONTINUED - delete the word " Plt." from the county column
ELECTIONS_DATA['County'] = ELECTIONS_DATA['County'].str.replace(' Plt.','')
ELECTIONS_DATA.head()
```

```python
## TASK 1 CONTINUED - merge the two datasets using inner join on the county and state and display the merged dataset
MERGED_DATA=pd.merge(ELECTIONS_DATA,DEMOGRAPHICS_DATA,how='inner',on=['County','State'])
MERGED_DATA
```

```python
# TASK 2 - explore the merged dataset and display the number of variables in the dataset
# (num1- rows, num2- variables/columns)
MERGED_DATA.shape
```

```python
# TASK 2 CONTINUED - dropping missing columns
MERGED_DATA.drop(['No Preference.Party','Uncommitted.Party'],axis=1, inplace=True)
MERGED_DATA.head()
```

```python
# TASK 2 CONTINUED - type of variables in the dataset
MERGED_DATA.info()
```

```python
# TASK 3 - create a new variable named “Democratic” that contains the number of votes cast for candidates from 
# the Democratic party in each county
MERGED_DATA['Democratic'] = MERGED_DATA['Bernie Sanders.Number of Votes']+MERGED_DATA['Hillary Clinton.Number of Votes']+MERGED_DATA['Martin O\'Malley.Number of Votes'] 
MERGED_DATA.head()
```

```python
# TASK 4 - create a new variable named “Republican” that contains the number of votes cast for candidates 
# from the Republican party in each county
MERGED_DATA['Republican'] = MERGED_DATA['Ben Carson.Number of Votes']+MERGED_DATA['Carly Fiorina.Number of Votes']+MERGED_DATA['Chris Christie.Number of Votes'] +MERGED_DATA['Donald Trump.Number of Votes'] +MERGED_DATA['Jeb Bush.Number of Votes'] +MERGED_DATA['John Kasich.Number of Votes'] +MERGED_DATA['Marco Rubio.Number of Votes'] +MERGED_DATA['Mike Huckabee.Number of Votes'] +MERGED_DATA['Rand Paul.Number of Votes'] +MERGED_DATA['Rick Santorum.Number of Votes']+MERGED_DATA['Ted Cruz.Number of Votes']
MERGED_DATA.head()
```

```python
# TASK 5 - create a new variable named “Party” that labels each county as Democratic or Republican. This new variable 
# should be equal to 1 if there were more votes cast for candidates from the Democratic party than the Republican
# party in that county and it should be equal to 0 otherwise
MERGED_DATA['Party'] = (MERGED_DATA['Democratic'] > MERGED_DATA['Republican']).astype(int)
MERGED_DATA.head()
```

```python
# TASK 6 - compute the mean population in 2014 for democratic counties
democraticPARTY = MERGED_DATA['Party'] == 1
democraticMEAN = MERGED_DATA.loc[democraticPARTY, '2014 Population'].mean()
democraticMEAN
```

```python
# TASK 6 CONTINUED - compute the mean population in 2014 for democratic counties
republicanPARTY = MERGED_DATA['Party'] == 0
republicanMEAN = MERGED_DATA.loc[republicanPARTY, '2014 Population'].mean()
republicanMEAN
```
